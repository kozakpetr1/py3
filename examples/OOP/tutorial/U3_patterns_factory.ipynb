```python
class Shape:
    def render(self):
        print("Initializing to render some shape...")

class Square(Shape):
    def render(self):
        super().render()
        print("Rendering square...")
        
class Triangle(Shape):
    def render(self):
        super().render()
        print("Rendering Triangle...")

class Circle(Shape):
    def render(self):
        super().render()
        print("Rendering Circle...")

class Factory:
    
    @staticmethod
    def create(representative):

        if representative == "Square":
            return Square()

        elif representative == "Triangle":
            return Triangle()

        elif representative == "Circle":
            return Circle()
        
        else:
            return Square()
        """ 
        match representative:
            
            case "Square":
                return Square()

            case "Triangle":
                return Triangle()

            case "Circle":
                return Circle()
                
            case _:
                return Square()
        """

my_square = Factory.create("Square") 
my_square.render()

my_triangle = Factory.create("Triangle")
my_triangle.render()

my_circle = Factory.create("Circle")
my_circle.render()
```

```python
class Sweets:
    pass

class Pralines(Sweets):
    
    def __init__(self, type_of):
        
        self.__type_of = type_of

class Truffles(Sweets):

    def __init__(self, type_of):

        self.__type_of = type_of
        
class Factory:
    
    def __init__(self):
        self.__sweets = []

    def showProducts(self):
        print(f"{self.__sweets}")

    def make(self, kind_of, type_of):
        
        if kind_of == "Pralines":
            self.__sweets.append(Pralines(type_of))
        
        elif kind_of == "Truffles":
            self.__sweets.append(Truffles(type_of))
            
sweets = Factory()
sweets.make("Pralines", "chocolate")
sweets.make("Truffles", "nougat")
sweets.make("Pralines", "salted caramel")
rum = Pralines("rum")
sweets.showProducts()
print(rum)

        
```

```python
class  Vehicle:
    pass

class SUV(Vehicle):
    
    def __init__(self, type_of):
        self.__type_of = type_of
        
    def showVehicle(self):
        print(self.__type_of)

class Sedan(Vehicle):

    def __init__(self, type_of):
        self.__type_of = type_of

class Coupe(Vehicle):

    def __init__(self, type_of):
        self.__type_of = type_of

class Factory:
    
    @staticmethod
    def make(kind_of, type_of):
    
        if kind_of == "SUV":
            return SUV(type_of)
            
        elif kind_of == "Sedan":
            return Sedan(type_of)
          
        elif kind_of == "Coupe":
            return Coupe(type_of)
         
        else:
            return False
        
v = []
v.append(Factory.make("SUV", "Ssangyong Korando"))
v.append(Factory.make("Sedan", "Škoda Superb"))
v.append(Factory.make("Coupe", "VW"))
v.append(Factory.make("SUV", "Škoda Karoq"))
v[0].showVehicle()
```

```python
class Darek:
    
    typDarku = ['Hracka', 'Zviratko', 'Obleceni', 'Domacnost', 'Doplnky', 'Sport']
    
    def __init__(self, typ_darku, jaky_darek):
        self.__typ = typ_darku
        self.__nazev = jaky_darek
        self.ukazDarek()
    
    def ukazDarek(self):
        print(self.__typ + ": " + self.__nazev + ".", end=" ")
        
    def __str__(self):
        return(f"{self.__typ}: {self.__nazev}") 
    
class Hracka(Darek):
    
    def __init__(self, typ_darku, jaky_darek):
        super().__init__(typ_darku, jaky_darek)
        print(f"Vítej v Království hraček. Vyrábím hračku {jaky_darek}.")     

class Zviratko(Darek):
    def __init__(self, typ_darku, jaky_darek):
        super().__init__(typ_darku, jaky_darek)
        print(f"Vítej v Ráji domácích mazlíčků. Připravuji zvířátko {jaky_darek}.")     

class Obleceni(Darek):
    def __init__(self, typ_darku, jaky_darek):
        super().__init__(typ_darku, jaky_darek)
        print(f"Vítej v Salonu měkkých dárků. Vyrábím {jaky_darek}.")     

class Domacnost(Darek):
    def __init__(self, typ_darku, jaky_darek):
        super().__init__(typ_darku, jaky_darek)
        print(f"Vítej v Oddělení pro domácnost. Vyrábím {jaky_darek}.")     

class Doplnky(Darek):
    def __init__(self, typ_darku, jaky_darek):
        super().__init__(typ_darku, jaky_darek)
        print(f"Vítej v Oddělení doplňků. Vyrábím {jaky_darek}.")     

class Sport(Darek):
    def __init__(self, typ_darku, jaky_darek):
        super().__init__(typ_darku, jaky_darek)
        print(f"Vítej v Oddělení pro sportovce. Vyrábím {jaky_darek}.")     

class Kosmetika(Darek):
    def __init__(self, typ_darku, jaky_darek):
        super().__init__(typ_darku, jaky_darek)
        print(f"Vítej v Oddělení kosmetiky. Vyrábím {jaky_darek}.")     

class SantovaTovarna:
    
    @staticmethod
    def vyrobitDarek(typ_hracky, nazev_hracky):
        if typ_hracky in Darek.typDarku:
            print("Vítej v Santově továrně.", end=" ")
            return globals()[typ_hracky](typ_hracky, nazev_hracky)
        else:
            return None

darek = []
darek.append(SantovaTovarna.vyrobitDarek('Hracka', 'Lego Friends'))
darek.append(SantovaTovarna.vyrobitDarek('Zviratko', 'Maltézský psík'))
darek.append(SantovaTovarna.vyrobitDarek('Obleceni', 'Ponožky'))
darek.append(SantovaTovarna.vyrobitDarek('Kosmetika', 'Rtěnka')) # toto nebude fungovat, Santova tovarna kosmetiku nevyrabi

darek.append(Kosmetika('Kosmetika', 'Rtěnka')) # vyroba darku mimo tovarnu
darek.append(Hracka('Hracka', 'Letadlo')) # vyroba darku mimo tovarnu

#for i in darek:
#    print(i)
    
```
