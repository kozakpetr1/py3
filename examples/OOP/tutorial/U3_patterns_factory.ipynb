```python
class Shape:
    def render(self):
        print("Initializing to render some shape...")

class Square(Shape):
    def render(self):
        super().render()
        print("Rendering square...")
        
class Triangle(Shape):
    def render(self):
        super().render()
        print("Rendering Triangle...")

class Circle(Shape):
    def render(self):
        super().render()
        print("Rendering Circle...")

class Factory:
    
    @staticmethod
    def create(representative):

        if representative == "Square":
            return Square()

        elif representative == "Triangle":
            return Triangle()

        elif representative == "Circle":
            return Circle()
        
        else:
            return Square()
        """ 
        match representative:
            
            case "Square":
                return Square()

            case "Triangle":
                return Triangle()

            case "Circle":
                return Circle()
                
            case _:
                return Square()
        """

f = Factory()
my_square = f.create("Square") 
my_square.render()

my_triangle = f.create("Triangle")
my_circle = f.create("Circle")
```

```python
class Sweets:
    pass

class Pralines(Sweets):
    
    def __init__(self, type_of):
        
        self.__type_of = type_of

class Truffles(Sweets):

    def __init__(self, type_of):

        self.__type_of = type_of
        
class Factory:
    
    def __init__(self):
        self.__sweets = []

    def showProducts(self):
        print(f"{self.__sweets}")

    def make(self, kind_of, type_of):
        
        if kind_of == "Pralines":
            self.__sweets.append(Pralines(type_of))
        
        elif kind_of == "Truffles":
            self.__sweets.append(Truffles(type_of))
            
f = Factory()
f.make("Pralines", "chocolate")
f.make("Truffles", "nougat")
f.make("Pralines", "salted caramel")
f.showProducts()


        

        
```

```python
class  Vehicle:
    pass

class SUV(Vehicle):
    
    def __init__(self, type_of):
        self.__type_of = type_of
        
    def showVehicle(self):
        print(self.__type_of)

class Sedan(Vehicle):

    def __init__(self, type_of):
        self.__type_of = type_of

class Coupe(Vehicle):

    def __init__(self, type_of):
        self.__type_of = type_of

class Factory:
    
    @staticmethod
    def make(kind_of, type_of):
    
        if kind_of == "SUV":
            return SUV(type_of)
            
        elif kind_of == "Sedan":
            return Sedan(type_of)
          
        elif kind_of == "Coupe":
            return Coupe(type_of)
         
        else:
            return False
        
f = Factory()

v = []
v.append(f.make("SUV", "Ssangyong Korando"))
v.append(f.make("Sedan", "Škoda Superb"))
v.append(f.make("Coupe", "VW"))
v.append(f.make("SUV", "Škoda Karoq"))
v[0].showVehicle()
```

```python
class Darek:
    
    typDarku = ['Hracka', 'Zviratko', 'Obleceni', 'Domacnost', 'Doplnky', 'Sport']
    
    def __init__(self, typ_darku, jaky_darek):
        self.__typ = typ_darku
        self.__nazev = jaky_darek
    
    def ukazDarek(self):
        print(self.__typ + ": " + self.__nazev)
        
    def __str__(self):
        return(f"{self.__typ}: {self.__nazev}") 
    
class Hracka(Darek):
    pass

class Zviratko(Darek):
    pass

class Obleceni(Darek):
    pass

class Domacnost(Darek):
    pass

class Doplnky(Darek):
    pass

class Sport(Darek):
    pass

class SantovaTovarna:
    
    @staticmethod
    def vyrobitDarek(typ_hracky, nazev_hracky):
        if typ_hracky in Darek.typDarku:
            return globals()[typ_hracky](typ_hracky, nazev_hracky)
        else:
            return None

Santa = SantovaTovarna()
darek = []
darek.append(Santa.vyrobitDarek('Hracka', 'Lego Friends'))
darek.append(Santa.vyrobitDarek('Zviratko', 'Maltézský psík'))
darek.append(Santa.vyrobitDarek('Obleceni', 'Ponožky'))
darek.append(Santa.vyrobitDarek('Kosmetika', 'Rtěnka'))

for i in darek:
    print(i)
    
```
